PYTHON CODE TO GENERATE A PIE CHART AND BAR CHART

```python
import matplotlib.pyplot as plt

# Data to plot
labels = ['MOMBASA', 'NAKURU', 'KISUMU']
sizes = [35, 30, 35]
colors = ['orange', 'green', 'blue']
explode = (0, 0, 0)  # explode a slice if required

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', startangle=200)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = ['FORD', 'BMW', 'AUDI', 'MERCEDES', 'JAGUAR', 'TESLA']
sizes = [15, 10, 20, 25, 10, 20]
colors = ['green', 'orange', 'blue', 'brown', 'purple', 'red']

# Plot
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=220)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

BAR CHART

```python
import matplotlib.pyplot as plt

# Data to plot
categories = ['SSET', 'Education', 'Health sci', 'Business', 'Law']
values = [25, 18, 35, 10, 30]

# Plot
plt.bar(categories, values, color='blue')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')

plt.show()
```
